```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils import resample
from sklearn.metrics import confusion_matrix
from sklearn.feature_selection import VarianceThreshold, SelectKBest, chi2, RFE
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score,recall_score,f1_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import pickle
```

```python
numerical=pd.read_csv("files_for_lab/numerical.csv")
categorical=pd.read_csv("files_for_lab/categorical.csv")
target=pd.read_csv("files_for_lab/target.csv")
display(numerical.head())
display(categorical.head())
display(target.head())
```

### Function Definitions

```python
def transform(X,y,random_state=0,verbose=False):
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random_state)
    
    X_train_num = X_train.select_dtypes(np.number)
    X_test_num = X_test.select_dtypes(np.number)

    X_train_cat = X_train.select_dtypes(object)
    X_test_cat = X_test.select_dtypes(object)

    transformer=MinMaxScaler().fit(X_train_num)
    
    X_train_num_scaled = transformer.transform(X_train_num)
    X_train_num_scaled = pd.DataFrame(X_train_num_scaled,columns=X_train_num.columns)

    X_test_num_scaled = transformer.transform(X_test_num)
    X_test_num_scaled = pd.DataFrame(X_test_num_scaled,columns=X_test_num.columns)

    encoder = OneHotEncoder().fit(X_train_cat)
    
    X_train_cat_encoded = encoder.transform(X_train_cat).toarray()
    cols = encoder.get_feature_names_out(input_features=X_train_cat.columns)
    X_train_cat_encoded = pd.DataFrame(X_train_cat_encoded, columns=cols)

    X_test_cat_encoded = encoder.transform(X_test_cat).toarray()
    cols = encoder.get_feature_names_out(input_features=X_test_cat.columns)
    X_test_cat_encoded = pd.DataFrame(X_test_cat_encoded, columns=cols)

    X_train = pd.concat([X_train_num_scaled, X_train_cat_encoded], axis=1)
    X_test = pd.concat([X_test_num_scaled, X_test_cat_encoded], axis=1)
    
    if verbose:
        print("X_train_num_scaled:")
        display(X_train_num_scaled.head())
        print("X_train_num_scaled:")
        display(X_test_num_scaled.head())
        print("X_train_cat_encoded:")
        display(X_train_cat_encoded.head())
        print("X_test_cat_encoded:")
        display(X_test_cat_encoded.head())
        print("X_train:")
        display(X_train.head())
        print("X_test:")
        display(X_test.head())
        
    return X_train, X_test, y_train, y_test, transformer, encoder
```

```python
def sample(X_train,y_train,output_colname,output_vals=[0,1],sample_state='down',verbose=False):
    train_set=pd.concat([X_train,y_train.reset_index(drop=True)],axis=1)
    
    if verbose:
        print("Initial Imbalance:")
        display(train_set[output_colname].value_counts())
    
    # This function ALWAYS presumes 0 category is dominant category in the dataset.
    category_0 = train_set[train_set[output_colname] == output_vals[0]]
    category_1 = train_set[train_set[output_colname] == output_vals[1]]
    
    if sample_state=='down':
        category_0_undersampled = resample(category_0, 
                                           replace=False, 
                                           n_samples = len(category_1))
        df_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
        
        if verbose:
            print("Imbalance after downsampling:")
            display(df_downsampled[output_colname].value_counts())
        
        df_downsampled = df_downsampled.sample(frac=1)
        X_train=df_downsampled.drop([output_colname],axis=1)
        
        if verbose:
            print("X_train:")
            display(X_train.head())
        
        y_train=df_downsampled[output_colname]
        
        if verbose:
            print("y_train:")
            display(y_train.head())
            
        return X_train, y_train
        
    elif sample_state=='up':
        category_1_oversampled = resample(category_1, 
                                          replace=True, 
                                          n_samples = len(category_0))

        df_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
        
        if verbose:
            print("Imbalance after upsampling:")
            display(df_upsampled[output_colname].value_counts())
        
        df_upsampled = df_upsampled.sample(frac=1)
        X_train=df_upsampled.drop([output_colname],axis=1)
        
        if verbose:
            print("X_train:")
            display(X_train.head())
            
        y_train=df_upsampled[output_colname]
        
        if verbose:
            print("y_train:")
            display(y_train.head())
            
        return X_train, y_train
    
    else:
        return None
```

```python
def disp_score_clf(clf, X_train, X_test, y_train, y_test):
    predictions = clf.predict(X_train)
    print("Training score: ",round(clf.score(X_train, y_train),2),"%")

    predictions = clf.predict(X_test)
    print("Test score: ",round(clf.score(X_test, y_test),2),"%")

    print("precision: ",round(precision_score(y_test,predictions),2))
    print("recall: ",round(recall_score(y_test,predictions),2))
    print("f1: ",round(f1_score(y_test,predictions),2))

    disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, predictions),display_labels=clf.classes_)
    disp.plot()
    plt.show()
```

```python
def disp_corr_heatmap(X,y,verbose=False):
    corr_matrix=pd.concat([X,y],axis=1).corr(method='pearson')
    if verbose:
        display(corr_matrix)
    fig, ax = plt.subplots(figsize=(len(X.columns), len(X.columns)))
    ax = sns.heatmap(corr_matrix, annot=True)
    plt.show()
```

The prediction score on the test data doesn't seem to be that good probably due to the fact that there are several columns that seem to be similar to each other contributing multicollinearity and thus low accuracy of the model. Therefore we'll go with a feature selction reducing the number of columns best fit for the model building and thereafter check for multicollinearity as well.

## Feature Selections

### Method 1: Variance Threshold

```python
scaler = MinMaxScaler()
numerical_scaled = scaler.fit_transform(numerical)

var_threshold = 0.02
sel = VarianceThreshold(threshold=(var_threshold))

sel = sel.fit(numerical_scaled)
selected_cols=[col[0] for col in zip(numerical.columns, list(sel.get_support())) if col[1] == True]
numerical_cleaned1=numerical[selected_cols]
numerical_cleaned1.head()
```

### Method 2: Select K Best

```python
X = numerical_scaled
y = target['TARGET_B']

kbest = SelectKBest(chi2, k=25).fit(X, y)
df = pd.DataFrame(data = kbest.scores_, columns = ['score'])
df['Column'] = numerical.columns
selected_cols=df.sort_values(by = ['score'], ascending = False).head(25)['Column']
numerical_cleaned2=numerical[selected_cols]
numerical_cleaned2.head()
```

### Method 3: Recursive Feature Elimination

```python
lm = linear_model.LinearRegression()
rfe = RFE(lm, n_features_to_select=25, verbose=False)
rfe.fit(X, y)

sel_cols_df = pd.DataFrame(data = rfe.ranking_, columns=['Rank'])
sel_cols_df['col_names'] = numerical.columns
sel_cols=list(sel_cols_df[sel_cols_df['Rank']<26]['col_names'])
numerical_cleaned3=numerical[sel_cols]
numerical_cleaned3.head()
```

### Identifying Multicollinearity among the features selected from SelectKBest

```python
disp_corr_heatmap(numerical_cleaned2,y,True)
```

On observing the correlation matrix for SelectKBest with top 25 columns, we still observe high multicollinearity between the columns, which requires us to remove several of them before we rebuild our column again.

As we can observe and interpet from the correlation matrix, there are multiple columns which are similar by their definition and have multiple variation columns contributing to multicollinearity. Removing those columns will surely mean losing of some information, but might improve the accuracy.

```python
X=pd.concat([numerical_cleaned2.drop(['ETH2','ETHC5','CARDPROM'], axis=1),categorical],axis=1)
y=target['TARGET_B']
X_train, X_test, y_train, y_test, transformer, encoder = transform(X,y,1143)
X_train, y_train = sample(X_train,y_train,'TARGET_B',[0,1],'up')

rfc = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.2,
                             random_state = 42)
rfc.fit(X_train, y_train)

disp_score_clf(rfc, X_train, X_test, y_train, y_test)

pickle.dump(rfc, open('rfc.p', 'wb'))
```

Even though our accuracy dropped further, but our recall increased by 5% which is significant. But, as our underlying accuracy isn't that great, we'll go for hyperparameter tuning which would help in increasing our accuracy.

### Improving the Hyperparameters to improve the accuracy of the model

```python
max_depth_choices= list(range(3,11))+[None]
criterion_choices = ['gini','entropy','log_loss']
min_samples_split_choices = list(range(2,11))
min_samples_leaf_choices = list(range(2,11))
max_samples_choices = [0.1,0.2,0.3]
max_features_choices = list(range(2,7))

grid = {'max_depth': max_depth_choices,
        'criterion': criterion_choices,
        'min_samples_split': min_samples_split_choices,
        'min_samples_leaf': min_samples_leaf_choices,
       'max_samples':max_samples_choices,
       'max_features':max_features_choices}


rfc = RandomForestClassifier()

grid_search = GridSearchCV(estimator = rfc, param_grid = grid, cv = 5)
grid_search.fit(X_train, y_train)
display(grid_search.best_params_)
display(round(grid_search.best_score_,2))
```

```python
rfc_params=grid_search.best_params_

rfc = RandomForestClassifier(max_depth=rfc_params['max_depth'],
                             min_samples_split=rfc_params['min_samples_split'],
                             min_samples_leaf =rfc_params['min_samples_leaf'],
                             max_samples=rfc_params['max_samples'],
                             random_state = rfc_params['max_features'])
rfc.fit(X_train, y_train)

disp_score_clf(rfc, X_train, X_test, y_train, y_test)

pickle.dump(rfc, open('rfc_v2.p', 'wb'))
```

If our objective is to maximize the net donations, it can be understood clearly by breaking the objective into 2 aspects:

+ First, we need to maximize our donation amounts, which further means that we correctly predict most of the potential donors.

+ Second, if we correctly predict the donors, then our mailing cost would go righly invested to gain maximum of the donations with minimal of the mailing cost in total (as each mail cost less than a $ or a euro.)

With an example, it goes like this:

If we are discovering the potential donors from a population of 1000, and from the actual donors of 180, we righly predict the donors as 150, miss 30 of them, but also wrongly send the donation requests to non potential donors, the net donations would be, considering the postal cost is 60 cents and minimum donation is 20$ (assuming):

||Predicted 0|Predicted 1|
|-|-|-|
|Actual 0|620|180|
|Actual 1|30|150|

Donation Amount GAINED by targeting the right donors (TP): (150*20)-(150*0.6) = 2910

Donation Amount lost by targeting the WRONG donors (FP): -(180*0.6) = -108

Net Donation Amount gained = 2910-108 = 2802

As we see, the false positive impact isn't that much, since the postal cost is little, therefore, our main driving factor is to minimize the False Negative, and the metric to optimize would be RECALL, which would significantly increase the overall donation amount.
